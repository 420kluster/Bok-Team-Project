```python
import os
import shutil

# ==========================================
# 1. 설정
SOURCE_DIR = "txt renamed"   # 현재 텍스트 파일들이 있는 폴더
TARGET_DIR = "BOK 금리 결정 의사록"  # 금리 결정 파일만 따로 모을 폴더
# ==========================================

# 금리 결정과 관련된 핵심 키워드
KEYWORDS = ["통화정책방향"]

if not os.path.exists(TARGET_DIR):
    os.makedirs(TARGET_DIR)

print(f">>> 금리 결정 관련 파일을 선별, '{TARGET_DIR}'로 복사\n")

files = [f for f in os.listdir(SOURCE_DIR) if f.lower().endswith(".txt")]
count = 0

for filename in files:
    file_path = os.path.join(SOURCE_DIR, filename)
    
    try:
        # 키워드 있는지 확인
        with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
            content = f.read()
            
            # 있으면 합격
            if any(key in content for key in KEYWORDS):
                # 새 폴더로 복사
                shutil.copy(file_path, os.path.join(TARGET_DIR, filename))
                print(f"발견: {filename}")
                count += 1
                
    except Exception as e:
        print(f"읽기 실패: {filename} ({e})")

print(f"\n>>> 완료! 총 {len(files)}개 중 {count}개의 파일을 찾았습니다.")
print(f">>> '{TARGET_DIR}' 폴더 확인")
```
